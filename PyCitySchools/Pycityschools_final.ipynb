```python
#import libarary dependencies 
import pandas as pd
from pathlib import Path

#set file path
schl_df = Path("Resources/schools_complete.csv")
stdnt_df = Path("Resources/students_complete.csv")

#set seperate objects for each csv.
schl_df = pd.read_csv(schl_df)
stdnt_df = pd.read_csv(stdnt_df)

schl_stdntdf = pd.merge(stdnt_df, schl_df, how="left", on=["school_name", "school_name"])
schl_stdntdf.head()
```

## __District Summary__ 

```python
# stdnt_df.head() / make sure both data sets read in 

# District Summary Metrics Portion 


#Calculate total number of unique schools
schl_df_unique = schl_stdntdf['school_name'].unique()
totalschools = len(schl_df_unique)
#print("Total Unique Schools ::",totalschools)



#Calculate total number of students 
stdnt_unique = schl_stdntdf['student_name'].value_counts()
total_stdnts = stdnt_unique.sum()
#print("Total Students ::",total_stdnts)

#calculate total district budget 
budget_ttl = schl_df['budget'].sum()
#print("District Budget Total ::",budget_ttl)

#Calculate District Ave. Math Score 
dis_avemathscr = schl_stdntdf['math_score'].mean()
dis_avemathscr = dis_avemathscr.round()
#print("District Ave. Math Score (%) ::",dis_avemathscr)

#Calculate District Ave. Reading Score 
dis_avereadscr = schl_stdntdf['reading_score'].mean()
dis_avereadscr = dis_avereadscr.round()
#print("District Ave. Reading Score (%) ::",dis_avereadscr)

#Calculate the % of studnets passing Math, Reading, and Overall just Passing.
lowmath = schl_stdntdf['math_score'].min() # = 55%
lowread = schl_stdntdf['reading_score'].min() # = 63%

#prcntpassmath = schl_stdntdf['math_score']>=70
#print(prcntpassmath)
#prcntpassmath[:,1].value_counts(normalize=True).mul(100).astype(str)+'%'
#prcntpassread = schl_stdntdf['reading_score']>=70
#for grade in stdnt_df[('math_score'),('reading_score')]:
    #if 
distpassmathcnt = schl_stdntdf[(schl_stdntdf["math_score"] >= 70)].count()["student_name"]
distpassmathpct = distpassmathcnt / float(total_stdnts) * 100
distpassmathpct = round(float(distpassmathpct), 2)
#print(distpassmathpct)

distpassreadcnt = schl_stdntdf[(schl_stdntdf["reading_score"] >= 70)].count()["student_name"]
distpassreadpct = distpassreadcnt / float(total_stdnts) * 100
distpassreadpct = round(float(distpassreadpct), 2)
#print(distpassreadpct)

prcntpassingcnt = schl_stdntdf[
    (schl_stdntdf["math_score"] >= 70) & 
    (schl_stdntdf["reading_score"] >= 70)
    ].count()["student_name"]

overall_passing = prcntpassingcnt / float(total_stdnts) * 100

overall_passing = round(float(overall_passing), 2)

dist_summary = pd.DataFrame({'Total Schools' : totalschools,
                             'Total Students' : total_stdnts,
                             'Total Budget' : budget_ttl,
                             'Ave. Math Score' : dis_avemathscr,
                             'Ave. Reading Score' : dis_avereadscr,
                             '% Passing Math' : distpassmathpct,
                             '% Passing Reading' : distpassreadpct,
                             '% Overall Passing' : overall_passing}, index=[0])


dist_summary["Total Students"] = dist_summary["Total Students"].map("{:,}".format)
dist_summary["Total Budget"] = dist_summary["Total Budget"].map("${:,.2f}".format)

dist_summary
```

## __School Summary__

```python
#per_stndt_buget = schl_df[['budget'/'size']]

schl_df['per_stndt_buget'] = schl_df['budget']/schl_df['size']
schl_df
```

```python
#Add school average math and reading scores to the summary table.
perschoolmathn = schl_stdntdf[['school_name', 'math_score']]
perschoolreadn = schl_stdntdf[['school_name', 'reading_score']]

perschoolmathn = perschoolmathn.groupby('school_name')
perschoolreadn = perschoolreadn.groupby('school_name')

schoolmathscor = perschoolmathn.mean()
schoolreadscor = perschoolreadn.mean()


schl_df = pd.merge(schl_df, schoolmathscor, on='school_name', how='left')
schl_df = pd.merge(schl_df, schoolreadscor, on='school_name', how='left')


schl_df
    
```

```python
#create an object that represents total stutdents that past math per school 
students_passing_math = schl_stdntdf[(schl_stdntdf["math_score"] >= 70)]
schlstndtspassmath = students_passing_math.groupby(["school_name"]).size()
schlstndtspassmath

#create an object that represents total stutdents that past reading per school 
students_passing_reading = schl_stdntdf[(schl_stdntdf["reading_score"] >= 70)]
schlstndtspassread = students_passing_reading.groupby(["school_name"]).size()
schlstndtspassread

#creat an object that represents total students that passed both math and reading per school
students_passing_overall = schl_stdntdf[(schl_stdntdf["math_score"] >= 70) & (schl_stdntdf["reading_score"] >= 70)]
schlstndtspassreadandmath = students_passing_overall.groupby(["school_name"]).size()
schlstndtspassreadandmath


#create divisible object that is the student size of each school
per_school_counts = schl_df.set_index(["school_name"])["size"]

#calculate the percent students that passed math, reading, overall math and reading per school
schlstndtspassmath = schlstndtspassmath / per_school_counts
schlstndtspassread = schlstndtspassread / per_school_counts
overall_passing_rate = schlstndtspassreadandmath / per_school_counts * 100

schlstndtspassmath = schlstndtspassmath * 100
schlstndtspassread = schlstndtspassread * 100

schlstndtspassmath = schlstndtspassmath.to_frame()
schlstndtspassread = schlstndtspassread.to_frame()
overall_passing_rate = overall_passing_rate.to_frame()

# Index of the 'name' column
columnindextorename = 0  
new_column_name = '% passed reading'
new_clm_name = '% passed math'
nuclmname = '% overall passing'

schlstndtspassmath.rename(columns={schlstndtspassmath.columns[columnindextorename]: new_clm_name}, inplace=True)
schlstndtspassread.rename(columns={schlstndtspassread.columns[columnindextorename]: new_column_name}, inplace=True)
overall_passing_rate.rename(columns={overall_passing_rate.columns[columnindextorename]: nuclmname}, inplace=True)


#merge the result series to the summary schl_df data frame. 
schl_df = pd.merge(schl_df, schlstndtspassmath, on='school_name', how='left')
schl_df = pd.merge(schl_df, schlstndtspassread, on='school_name', how='left')
schl_df = pd.merge(schl_df, overall_passing_rate, on='school_name', how='left')


schl_summary = schl_df.rename(columns={'math_score': 'Ave. Math Score', 
                                       'reading_score': 'Ave. Reading Score',
                                       'per_stndt_buget' : 'Per Student Budget',
                                       'budget' : 'Total School Budget'})

schl_summary["Total School Budget"] = schl_summary["Total School Budget"].map("${:,.2f}".format)
schl_summary["Per Student Budget"] = schl_summary["Per Student Budget"].map("${:,.2f}".format)




schl_summary
```

## __Math Scores by Grade__

```python
#Calculating Math Scores By Grade per school

ninth_graders = schl_stdntdf[(schl_stdntdf["grade"] == "9th")]
tenth_graders = schl_stdntdf[(schl_stdntdf["grade"] == "10th")]
eleventh_graders = schl_stdntdf[(schl_stdntdf["grade"] == "11th")]
twelfth_graders = schl_stdntdf[(schl_stdntdf["grade"] == "12th")]


ninth_grade_math_scores = ninth_graders.groupby(['school_name'])['math_score'].mean()
tenth_grade_math_scores = tenth_graders.groupby(['school_name'])['math_score'].mean()
elevnth_grade_math_scores = eleventh_graders.groupby(['school_name'])['math_score'].mean()
twelf_grade_math_scores = twelfth_graders.groupby(['school_name'])['math_score'].mean()

ninth_grade_math_scores = ninth_grade_math_scores.to_frame()
tenth_grade_math_scores = tenth_grade_math_scores.to_frame()
elevnth_grade_math_scores = elevnth_grade_math_scores.to_frame()
twelf_grade_math_scores = twelf_grade_math_scores.to_frame()


columnindextorename = 0  
new_column_name = '9th Grade Math Score'
new_clm_name = '10th Grade Math Score'
nuclmname = '11th Grade Math Score'
nclmnm = '12th Grade Math Score'

ninth_grade_math_scores.rename(columns={ninth_grade_math_scores.columns[columnindextorename]: new_column_name}, inplace=True)
tenth_grade_math_scores.rename(columns={tenth_grade_math_scores.columns[columnindextorename]: new_clm_name}, inplace=True)
elevnth_grade_math_scores.rename(columns={elevnth_grade_math_scores.columns[columnindextorename]: nuclmname}, inplace=True)
twelf_grade_math_scores.rename(columns={twelf_grade_math_scores.columns[columnindextorename]: nclmnm}, inplace=True)

#merge the result series to the summary schl_df data frame. 
math_score_by_grade = pd.merge(ninth_grade_math_scores, tenth_grade_math_scores, on='school_name', how='left')
math_score_by_grade = pd.merge(math_score_by_grade, elevnth_grade_math_scores, on='school_name', how='left')
math_score_by_grade = pd.merge(math_score_by_grade, twelf_grade_math_scores, on='school_name', how='left')
math_score_by_grade
```

##  __Scores by School Spending (per student)__

```python
#School performance based on spending per studnet budget. 

#establish spending bins 
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]

```

```python
#create copy of school summary data frame as it has the metrics we need (per student budget).
school_spending_df = schl_summary.copy()

school_spending_df[school_spending_df.columns[3:5]] = school_spending_df[school_spending_df.columns[3:5]].replace('[\$,]', '', regex=True).astype(float)
```

```python
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_spending_df['Per Student Budget'], bins = spending_bins, labels = labels, include_lowest=True)

school_spending_df["Total School Budget"] = school_spending_df["Total School Budget"].map("${:,.2f}".format)
school_spending_df["Per Student Budget"] = school_spending_df["Per Student Budget"].map("${:,.2f}".format)

school_spending_df
```

```python
#per student spending summary on academic success.

spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Ave. Math Score"].mean()
spending_math_scores = spending_math_scores.to_frame()
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Ave. Reading Score"].mean()
spending_reading_scores = spending_reading_scores.to_frame()
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% passed math"].mean()
spending_passing_math = spending_passing_math.to_frame()
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% passed reading"].mean()
spending_passing_reading = spending_passing_reading.to_frame()
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% overall passing"].mean()
overall_passing_spending = overall_passing_spending.to_frame()
```

```python
spending_summary = pd.merge(spending_math_scores, spending_reading_scores, on='Spending Ranges (Per Student)', how='left')
spending_summary = pd.merge(spending_summary, spending_passing_math, on='Spending Ranges (Per Student)', how='left')
spending_summary = pd.merge(spending_summary, spending_passing_reading, on='Spending Ranges (Per Student)', how='left')
spending_summary = pd.merge(spending_summary, overall_passing_spending, on='Spending Ranges (Per Student)', how='left')
spending_summary
```

##  __Scores by School Size__

```python
#Scores by school size summary 

# Establish the bins.
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
schl_summary["School Size"] = pd.cut(schl_summary["size"], bins = size_bins, labels = labels, include_lowest=True)
schl_summary
```

```python
#Per School Size affect on Academic succsess.

size_math_scores = schl_summary.groupby(["School Size"])["Ave. Math Score"].mean()
size_math_scores = size_math_scores.to_frame()
size_reading_scores = schl_summary.groupby(["School Size"])["Ave. Reading Score"].mean()
size_reading_scores = size_reading_scores.to_frame()
size_passing_math = schl_summary.groupby(["School Size"])["% passed math"].mean()
size_passing_math = size_passing_math.to_frame()
size_passing_reading = schl_summary.groupby(["School Size"])["% passed reading"].mean()
size_passing_reading= size_passing_reading.to_frame()
size_overall_passing = schl_summary.groupby(["School Size"])["% overall passing"].mean()
size_overall_passing = size_overall_passing.to_frame()
```

```python
size_summary = pd.merge(size_math_scores, size_reading_scores, on='School Size', how='left')
size_summary = pd.merge(size_summary, size_passing_math, on='School Size', how='left')
size_summary = pd.merge(size_summary, size_passing_reading, on='School Size', how='left')
size_summary = pd.merge(size_summary, size_overall_passing, on='School Size', how='left')
size_summary
```
